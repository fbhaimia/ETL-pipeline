# Cleaning Disaster Messages Dataset

```python
# checking for null values
round(messages.isnull().sum()*100/len(messages),2)
```

```python
# dropping original column from dataset 1
messages_cld = messages.drop('original', axis = 1)
messages_cld.columns
```

# Cleaning Disaster Categories Dataset

```python
# spliting categories column with delimiter
categories_cld = categories["categories"].str.split(";", expand=True)
categories_cld.head()
```

```python
# renaming columns
row = categories_cld.head(1)
category_colnames = row.applymap(lambda x: x[:-2]).iloc[0, :].tolist()
categories_cld.columns = category_colnames
categories_cld.head()
```

# Encoding

```python
# Encoding 
for column in categories_cld:
    categories_cld[column] = categories_cld[column].astype(str).str[-1] #setting value as per the last character of string
    categories[column] = categories_cld[column].astype(int) #converting to int
categories_cld.head()
```

# Merging Dataset

```python
# merging category ID with cleaned cleaned categories dataset
categories_id = categories['id']
categories_id = pd.DataFrame(categories_id)
categories_cld = pd.DataFrame(categories_cld)
categories_cld_df = pd.concat([categories_id, categories_cld], axis=1)
categories_cld_df.head()
```

```python
# merging both the datasets with ID.
df_merged = messages_cld.merge(categories_cld_df, on=["id"])
df_merged.head()
```

# Duplicates Treatment

```python
# shape of dataset before removing duplicates
df_merged.shape
```

```python
# duplicate values
df_merged[df_merged.duplicated()]
```

```python
# dropping duplicate values
df_merged.drop_duplicates(inplace = True)
```

```python
# shape of dataset after removing duplicates
df_merged.shape
```

# Load

1. Loading the processsed dataframe by creating the database.
2. Reading the dataframe from SQL database.
3. Closing the database connection

```python
# loading data into database
conn = sqlite3.connect('diaster.db')
df_merged.to_sql('disaster_processed', con = conn, if_exists='replace', index=False)
```

```python
# reading data from database
pd.read_sql('SELECT * FROM disaster_processed', con = conn).head()
```

```python
# closing connection
conn.commit()
conn.close()
```
